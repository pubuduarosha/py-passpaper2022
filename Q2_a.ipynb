#Csv to txt 

import pandas as pd

def convert_csv_to_text(csv_file, text_file):
  """Converts a CSV file to text.

  Args:
    csv_file: The path to the CSV file.
    text_file: The path to the text file.
  """

  df = pd.read_csv(csv_file)
  text = df.to_string()

  with open(text_file, "w") as f:
    f.write(text)

if __name__ == "__main__":
  csv_file = "data/Q2.csv"
  text_file = "output.txt"

  convert_csv_to_text(csv_file, text_file)

```python
# Temperature Distribution
plt.figure(figsize=(12, 6))
sns.histplot(data['TempHighF'], kde=True, label='TempHighF', color='red')
sns.histplot(data['TempAvgF'], kde=True, label='TempAvgF', color='blue')
sns.histplot(data['TempLowF'], kde=True, label='TempLowF', color='green')
plt.xlabel('Temperature (Fahrenheit)')
plt.title('Temperature Distribution')
plt.legend()
plt.show()
```

```python
# Humidity Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data['HumidityHighPercent'], kde=True, label='HumidityHighPercent', color='red')
sns.histplot(data['HumidityAvgPercent'], kde=True, label='HumidityAvgPercent', color='blue')
sns.histplot(data['HumidityLowPercent'], kde=True, label='HumidityLowPercent', color='green')
plt.xlabel('Humidity (%)')
plt.title('Humidity Distribution')
plt.legend()
plt.show()
```

```python
# Precipitation Analysis
plt.figure(figsize=(12, 6))
sns.histplot(data['PrecipitationSumInches'], kde=True, color='blue')
plt.xlabel('Precipitation (Inches)')
plt.title('Precipitation Distribution')
plt.show()
```

```python
# Precipitation Events
plt.figure(figsize=(12, 6))
sns.countplot(y=data['Events'], order=data['Events'].value_counts().index)
plt.xlabel('Count')
plt.title('Precipitation Events')
plt.show()
```

```python
# Sea Level Pressure Analysis
plt.figure(figsize=(12, 6))
sns.histplot(data['SeaLevelPressureHighInches'], kde=True, label='High', color='red')
sns.histplot(data['SeaLevelPressureAvgInches'], kde=True, label='Average', color='blue')
sns.histplot(data['SeaLevelPressureLowInches'], kde=True, label='Low', color='green')
plt.xlabel('Sea Level Pressure (Inches)')
plt.title('Sea Level Pressure Distribution')
plt.legend()
plt.show()
```

```python
#Box Plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['TempHighF', 'TempAvgF', 'TempLowF']], palette='Set2')
plt.title('Temperature Box Plots')
plt.ylabel('Temperature (Â°F)')
plt.show()
```

Findings:

The box plots show the variability in daily temperatures.

"TempHighF" tends to be higher than "TempAvgF" and "TempLowF."

There are some outliers in "TempHighF," indicating extremely high temperatures on some days.

```python
#Correlation Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Findings:

There is a strong positive correlation between "TempHighF" and "TempAvgF"

There is a moderate negative correlation between "TempHighF" and "PrecipitationSumInches," indicating that higher temperatures are associated with lower precipitation

```python
#Precipitation Analysis
plt.figure(figsize=(10, 5))
sns.histplot(data['PrecipitationSumInches'], kde=True, color='blue')
plt.xlabel('Precipitation (Inches)')
plt.title('Precipitation Distribution')
plt.show()
```

Findings:

Most days have low precipitation, but there are days with higher values.

The distribution is right-skewed, indicating that the majority of days have little to no precipitation.
